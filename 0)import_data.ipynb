```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dbs = yf.Ticker("DBSDF")
uob = yf.Ticker("UOVEY")
```

```python
start_date = "2017-04-18"
end_date = "2025-04-18"
dbs.history(start=start_date,end=end_date).plot(y='Close',ylabel="DBSDF Price", title="DBSDF Price History")
```

```python
uob.history(start=start_date,end=end_date).plot(y='Close',ylabel="UOVEY Price", title="UOVEY Price History")
```

```python
dbs_history = dbs.history(start=start_date, end=end_date)
uob_history = uob.history(start=start_date, end=end_date)
dbs_history.index = dbs_history.index.strftime('%Y-%m-%d')
uob_history.index = uob_history.index.strftime('%Y-%m-%d')

dbs_history.to_csv("dbs_7y.csv")
uob_history.to_csv("uob_7y.csv")
```
